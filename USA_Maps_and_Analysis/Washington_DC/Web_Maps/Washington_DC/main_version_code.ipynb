```python

import folium
```

```python
# Create a Map:
map = folium.Map(
    location = [37, 0],
    zoom_start = 2.5,
    tiles = 'cartodb positron',
)
```

```python
# GeoJSON file sources:
geojson_wards_file = "./Data/Wards_from_2022.geojson"
geojson_hospitals_file = "./Data/Hospitals.geojson"
geojson_libraries_file = "./Data/Libraries.geojson"
geojson_marinas_file = "./Data/Marinas.geojson"
geojson_national_parks_file = "./Data/National_Parks.geojson"
geojson_police_stations_file = "./Data/Police_Stations.geojson"
geojson_post_offices_file = "./Data/Post_Offices.geojson"
```

```python
# GeoJSON variable:
geojson_wards = folium.GeoJson(
    geojson_wards_file,
    name = "Boundaries"
).add_to(map)


# Display Hospitals:
geojson_hospitals = folium.GeoJson(
    geojson_hospitals_file,
    name = 'Hospitals',
    
    # Add Marker:
    marker = folium.Marker(
        icon=folium.Icon(icon_color = "red", icon = "h-square", prefix='fa')
    ),
    
    # Add tooltip to data (display data on hover):
    tooltip = folium.GeoJsonTooltip(aliases = ["Hospital Name", "Hospital Address"], fields = ["NAME", "ADDRESS"])
    
).add_to(map)


# Display Libraries:
geojson_libraries = folium.GeoJson(
    geojson_libraries_file,
    name = "Libraries",
    
    # Add Marker:
    marker = folium.Marker(
        icon=folium.Icon(icon = "book", prefix='fa')
    ),
    
    # Add tooltip to data (display data on hover):
    tooltip = folium.GeoJsonTooltip(aliases = ["Library Name", "Library Address"], fields = ["NAME", "ADDRESS"])
).add_to(map)



# Display Marinas:
geojson_marinas = folium.GeoJson(
    geojson_marinas_file,
    name = "Marinas",
    
    # Add Marker:
    marker = folium.Marker(
        icon=folium.Icon(icon = "ship", prefix='fa')
    ),
    
    # Add tooltip to data (display data on hover):
    tooltip = folium.GeoJsonTooltip(aliases = ["Marina's Name", "Marina's Address"], fields = ["NAME", "ADDRESS"])
).add_to(map)


# Display National Parks:
geojson_national_parks = folium.GeoJson(
    geojson_national_parks_file,
    name = "National Parks",
    
    # Add Marker:
    marker = folium.Marker(
        color = 'green',
        icon = folium.Icon(icon = 'tree', prefix = 'fa')
    ),
    
    # Add tooltip to data (display data on hover):
    tooltip = folium.GeoJsonTooltip(aliases = ["Park Name", "Park Code"], fields = ["NAME", "PARK_CODE"])
    
).add_to(map)


# Display Police Stations:
geojson_police_stations = folium.GeoJson(
    geojson_police_stations_file,
    name = "Police Stations",
    
    # Add Marker:
    marker = folium.Marker(
        color = 'green',
        icon = folium.Icon(icon = 'building', prefix = 'fa')
    ),
    
    # Add tooltip to data (display data on hover):
    tooltip = folium.GeoJsonTooltip(aliases = ["Police Station Name"], fields = ["NAME"])
).add_to(map)



# Display Post Offices:
geojson_post_offices= folium.GeoJson(
    geojson_post_offices_file,
    name = "Post Offices",
    
    # Add Marker:
    marker = folium.Marker(
        icon = folium.Icon(icon = 'envelope', prefix = 'fa')
    ),
    
    # Add tooltip to data (display data on hover):
    tooltip = folium.GeoJsonTooltip(aliases = ["Post Office Name", "Post Office Address"], fields = ["NAME", "ADDRESS"])
).add_to(map) 


# Add a Layer Control to the map
folium.LayerControl().add_to(map)
```
